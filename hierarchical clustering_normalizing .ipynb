```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import normalize
datascaled=pd.DataFrame(normalize(data),columns=data.columns)
```

```python
datascaled
```

```python
import scipy.cluster.hierarchy as hi
plt.figure(figsize=(10,5))
den=hi.dendrogram(hi.linkage(datascaled,method='ward'))
plt.axhline(y=6,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering()
model.fit_predict(datascaled)
```

```python
plt.scatter(datascaled['Milk'],datascaled['Grocery'],c=model.labels_)
```

```python
data['labels']=model.labels_
data.groupby('labels').agg('mean')
```
